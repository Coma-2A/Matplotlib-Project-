### Grouped Bar plots
create a grouped bar plot of the data below:
<br> Indicate the ff
- label
- title
- bar_label
- xticks
- legend

```python
#data
countries=['Ghana','Portugal','South Korea','Uruguay','Germany']
gold=[46,27,26,19,17]
silver=[37,23,18,18,10]
bronze=[38,30,25,18,24]
```

Solution of Grouped Bar plots

```python
x=['Ghana','Portugal','South Korea','Uruguay','Germany']
w=0.2
bar1=np.arange(len(x))
bar2=[i+w for i in bar1]
bar3=[i+w for i in bar2]

plt.bar(bar1,gold,w,label='Gold',color='red')
plt.bar(bar2,silver,w,label='Silver',color='blue')
plt.bar(bar3,bronze,w,label='Bronze',color='green')



plt.ylabel('Number of Trophy Won')
plt.xticks(bar1+w,x)
plt.legend()
plt.title('Trophy number and type won by 5 countries in Olympics 2016')
plt.show()
```
